### 1-Sınavlardaki performans cinsiyete, kökene ve anne-baba eğitim durumuna göre değişiklik göstermekte midir?

#### Burada karşılaştıracağımız veriler sürekli("Sınav performansı") ve kategorik("Köken,cinsiyet,anne-baba eğitim düzeyi") veriler olduğu için bar grafiğini kullanalım.

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df
```

```python
df.info()
```

```python
#Math score daki aşırı değerlere bakalım
sns.boxplot(x = df["math score"])
#Null değerleri inceleyelim
df["math score"].isnull().sum()
```

```python
#Bulduğumuz aykırı değerleri winsorize edelim
from scipy.stats.mstats import winsorize
winsorize_math = winsorize(df["math score"],(0, 0.10))
sns.boxplot(x = winsorize_math)
plt.title("Winsorized math")
```

```python
#0 ve - değerleri 1 yaparak, log alalım winsorize edelim
a=0
for i in df["math score"]:
    if i == 0 or i<0:
        df["math score"][a] = 1
    a+=1
```

```python
sns.boxplot(x = np.log(df["math score"]))
```

```python
winsorize_math = winsorize(df["math score"],(0, 0.10))
sns.boxplot(x = winsorize_math)
```

```python
new = ["race/ethnicity","gender"]
plt.figure(figsize=(20,5))
for i in new:
    plt.subplot(1,2,new.index(i)+1)
    sns.barplot(x=df[i],y= df["math score"])
    

plt.figure(figsize=(20,5))
sns.barplot(x=df["parental level of education"],y=df["math score"])
```

```python
#Yukarıdaki bar grafiklere göre farklılılıkları görebiliyoruz.
```

## Birbiriyle korelasyonu en fazla olan dersler hangisidir?

```python
#İki sürekli değişkene bakıcağımız için burada scatter grafiği kullanalım
lessons = ["math score","reading score","writing score"]
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.scatter(df["math score"],df["reading score"])
plt.xlabel("Math score")
plt.ylabel("Reading score")

plt.subplot(1,3,2)
plt.scatter(df["math score"],df["writing score"])
plt.xlabel("Math score")
plt.ylabel("Writing score")

plt.subplot(1,3,3)
plt.scatter(df["reading score"],df["writing score"])
plt.xlabel("Reading score")
plt.ylabel("Writing score")
```

```python
#Korelasyonları inceleyelim
df.corr()
```
