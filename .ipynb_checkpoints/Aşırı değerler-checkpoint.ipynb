## Öncelikle bir önceki derste uyguladağınız eksik değerleri tamamlama tekniklerinden en uygun olanını uygulayın. Aşağıdaki sorulara cevap verebilmeniz için eksik verilerin üstesinden gelmelisiniz.

### 1-Toplam gelir verilerini içeren TOTAL_REVENUE ve toplam harcama verilerini içeren TOTAL_EXPENDITURE değişkenlerini inceleyin. Bu değişkenler aykırı değerler içeriyor mu?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("states_all.csv")
```

```python
df.columns
```

```python
df.info()
```

```python
#Verimizdeki null değerleri ortalamaya göre dolduralım
for i in df.columns:
    print(df[i].dtype)
    if df[i].dtype == "int64" or df[i].dtype == "float64":
        df[i].fillna(df[i].mean(),inplace=True)
```

```python
#Null değerimiz var mı kontrol edelim
df.isnull().sum()
```

```python
#Aykırı değerlere bakmak için kutu grafiğini inceleyelim

plt.figure(figsize=(10,5),dpi=200)

plt.subplot(1,2,1)
plt.title("Toplam harcama kutu grafiği")
plt.boxplot(df.TOTAL_EXPENDITURE)

plt.subplot(1,2,2)
plt.title("Toplam kazanç kutu grafiği")
plt.boxplot(df.TOTAL_REVENUE)
plt.show()
```

### Kutu grafiklerimize baktığımızda,her iki değişken de aykırı değerler içeriyor

```python
#Bu sefer de aykırı değerlere z-score ile bakalım
from scipy.stats import zscore
for i in df.columns:
    if df[i].dtype == "int64" or df[i].dtype == "float64":
        z_score_new = zscore(df[i])
        for threshold in range(1,5):
            print("{} colonu için {} eşik değerinde aykırı değer sayısı {}".format(i,threshold,len(np.where(z_score_new > threshold)[0])))
```

```python
#Aykırı değerleri tukey yöntemi ile bulalım
esik_degerleri = pd.DataFrame()
for i in df.columns:
    if df[i].dtype == "int64" or df[i].dtype == "float64":
        q75, q25 = np.percentile(df[i], [75 ,25])
        caa = q75 - q25
        for threshold in np.arange(1,5,0.5):
            min_deger = q25 - (caa*threshold)
            maks_deger = q75 + (caa*threshold)
            aykırı_değerler = len((np.where((df[i] > maks_deger)|(df[i] < min_deger))[0]))
            #print("{} colonu için {} eşik değerinde aykırı değer sayısı {}".format(i,threshold,aykırı_değerler))
            esik_degerleri = esik_degerleri.append({'esik_degeri': threshold,'aykiri_deger_sayısı': aykırı_değerler}, 
                                           ignore_index=True)
esik_degerleri
```

### 2-Toplam gelir (TOTAL_REVENUE) ve toplam harcama (TOTAL_EXPENDITURE) değişkenlerinde aykırı değer tespit ettiyseniz bu derste öğrendiğiniz teknikleri uygulayarak bu aykırı değerleri ortadan kaldırın ve bunu yaptıktan sonra hiçbir aykırı değer kalmadığını doğrulayın.

### 1- Aykırı değerleri veri kümemizden çıkarabiliriz.

### 2- Aykırı değerleri sınırlandırabiliriz

```python
#Winsorize etme yöntemini kullanalım, %90 lık dilimi kullanıyoruz
from scipy.stats.mstats import winsorize
wins_data = winsorize(df.TOTAL_REVENUE, (0, 0.10))

#Kutu grafiğinde winsorize edilmiş datamıza bakalım
plt.boxplot(wins_data)

#Before-after karşılaştırma yapalım

plt.figure(figsize=(10,5),dpi=200)

plt.subplot(1,2,1)
plt.boxplot(df.TOTAL_REVENUE)
plt.title("Normal data")

plt.subplot(1,2,2)
plt.boxplot(wins_data)
plt.title("Winsorized data")

plt.show()
```

### 3-Aykırı değerleri dönüştürme

```python
#Karekök ve log dönüşümü yaparak ayrıkı değerlerden kurtulabiliriz, log dönüşümü yapalım

plt.figure(figsize=(10,5),dpi=200)

plt.subplot(1,2,1)
plt.boxplot(df.TOTAL_REVENUE)
plt.title("Normal data")

plt.subplot(1,2,2)
plt.boxplot(np.log(df.TOTAL_REVENUE))
plt.title("Logged data")

plt.show()
```

## 3-Toplam gelir (TOTAL_REVENUE) değişkeninden, toplam harcama (TOTAL_EXPENDITURE) değerini çıkartarak bütçe açığı olarak düşünülebilecek bir değişken oluşturun (aykırı değerleri temezlemeden). Bu yeni değişkende de aykırı değerler yer almakta mı? Varsa bu aykırı değerleri de temizleyin.

```python
df = pd.read_csv("states_all.csv")
```

```python
df
```

```python
df["Bütçe_açığı"] = df.TOTAL_REVENUE - df.TOTAL_EXPENDITURE 
```

```python
df
```

```python
#Bütçe açığında null değerleri dolduralım
df.Bütçe_açığı.fillna(df.Bütçe_açığı.mean(),inplace=True)

#Bütçe açığında aykırı değerlere bakalım
plt.boxplot(df.Bütçe_açığı)
plt.show()

#Bütçe açığındaki aykırı değerlier görebiliyoruz
```

```python
df.Bütçe_açığı.isnull().sum()
```

```python
df.Bütçe_açığı
```

```python
#Winsorize edelim ve karşılaştıralım
plt.figure(figsize = (10,5),dpi=200)

plt.subplot(1,2,1)
plt.boxplot(df.Bütçe_açığı)
plt.title("Normal bütçe açığı")

plt.subplot(1,2,2)
plt.boxplot(winsorize(df.Bütçe_açığı,(0.05, 0.05)))
plt.title("Winsorize edilmiş bütçe açığı")

plt.show
```

## 4-Yine toplam gelir (TOTAL_REVENUE) değişkeninden, toplam harcama(TOTAL_EXPENDITURE) değerini çıkartarak bütçe açığı olarak düşünülebilecek bir değişken oluşturun. Fakat bu sefer aykırı değerleri temezledikten sonra bu işlemi yapın. Bu yeni değişkende de aykırı değerler var mı? Varsa bunları da temizlemekte fayda var.

```python
df.TOTAL_REVENUE.fillna(df.TOTAL_REVENUE.mean(),inplace=True)
df.TOTAL_EXPENDITURE.fillna(df.TOTAL_REVENUE.mean(),inplace=True)
wins_data_tr = winsorize(df.TOTAL_REVENUE, (0, 0.10))
wins_data_te = winsorize(df.TOTAL_EXPENDITURE, (0, 0.10))
df.TOTAL_REVENUE = wins_data_tr
df.TOTAL_EXPENDITURE = wins_data_te
df["Bütçe_açığı"] = df.TOTAL_REVENUE - df.TOTAL_EXPENDITURE
plt.boxplot(df["Bütçe_açığı"])
plt.title("Winsorized datalardan elde edilen bütçe açığı")
plt.show()
```
