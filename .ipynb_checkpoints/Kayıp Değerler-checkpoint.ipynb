## 1-Veri kümesinde yer alan değişkenlerinin tiplerini ve her bir değişkenin eksik (null) değer oranını bulun.

```python
df = pd.read_csv("states_all.csv")
```

```python
df
```

```python
df.info()
```

```python
df.isnull().head()
```

```python
# Tüm sütunlardaki eksik değerlerin yüzdelerini hesaplayalım
df.isnull().sum()*100/df.shape[0]
```

## 2- Verimizde yıl (year) sütunu olduğunu farketmişsinizdir. Şimdilik yıl verisini unutun ve her bir gözlemin aynı yıl içerisinde yapıldığını farz edin. Her bir değişken için eksik değerleri nasıl doldurabileceğinizi düşünün. Eksik değerleri bir değerle doldurmak hangi değişkenler için anlamlı, hangileri için anlamsızdır?

```python
if df["YEAR"].dtype == "int64":
    print(True)
```

```python
#Sürekli değişkenlere ait boş verileri ortalama alarak dolduralım
for i in df.columns:
    if df[i].dtype == "int64" or df[i].dtype == "float64":
        df[i].fillna(df[i].mean(),inplace=True)
```

```python
df
```

```python
#Yeniden tüm sütunlardaki eksik değerlerin oranını hesaplayalım
df.isnull().sum()*100/df.shape[0]
```

## Tüm eksik değerleri doldurduk

```python
# Her bir sütuna ait tüm benzersiz değerleri alalım
for column in df.columns:
    print("{} sütunundaki benzersiz değerler : {}".format(column, df[column].unique()))
```

## 3- Şimdi zaman faktörünü dikkate alma zamanı! 2. sorudaki cevabınızı tekrar gözden geçirin ve eksik verileri o yıl içerisinde gözlemlenen değerlere dayanarak doldurun. Örneğin, bir değeri ortalama değer ile doldurmak isterseniz, o yılın ortalamasını hesaplayın.

```python
#Veriyi yeniden okuyalım
df = pd.read_csv("states_all.csv")
```

```python
df
```

```python
new_df = df.loc[lambda df: df["YEAR"] == 1992]
new_df.mean()
```

```python
df.groupby(["YEAR"]).mean("ENROLL")
```

```python
for i in df.columns:
    if df[i].dtype == "int64" or df[i].dtype == "float64":
        #O yıla ait ortalamayı hesaplayalım
        new_df = df.loc[lambda df: df["YEAR"] == df.YEAR]
        df[i].fillna(new_df[i].mean(),inplace=True)
```
